```python
from okx.api.market import Market

flag = "0"  # :0 , 1

marketDataAPI =  Market()

# 
result = marketDataAPI.get_tickers(
    instType="SPOT"
)

print(result)
```

```python
inst_ids = [item['instId'] for item in result['data']]
print(inst_ids)
```

```python
len(inst_ids)
```

```python
# 
pairs = {}

# inst_ids
for inst_id in inst_ids:
    # yy
    token, stable = inst_id.split('-')
    # inst_id
    if stable in pairs:
        pairs[stable].append(token)
    else:
        pairs[stable] = [token]

pairs = {k: sorted(v) for k, v in pairs.items()}

# 
for category, values in pairs.items():
    print(f"Pairs {category}: {values}")
```

```python
def get_week_closing_price(instId: str):
    week_closing_price = {}
    candle_data = marketDataAPI.get_history_candles(instId, bar='1Wutc')

    print(candle_data)

    for item in candle_data['data']:
        week_closing_price[item[0]] = item[4]

    return week_closing_price
```

```python
week_closing_price = []
candle_data = marketDataAPI.get_history_candles('BTC-USDT', bar='1Wutc')

for item in candle_data['data']:
        week_closing_price.append([item[0], item[4]])

print(candle_data)
print(week_closing_price[-1][0])
```
