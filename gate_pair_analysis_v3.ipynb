```python
# coding: utf-8
import requests
import time
import pandas as pd
import matplotlib.pyplot as plt

# 绘制滤波前后的时间序列图
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为宋体或其他已安装的字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

PATH = 'D:/All About Learning/Quant/gate/'
```

```python
def get_candle_sticks(pair: str, after: str = '', before: str = '', interval: str = '7d'):
    host = "https://api.gateio.ws"
    prefix = "/api/v4"
    headers = {'Accept': 'application/json', 'Content-Type': 'application/json'}

    url = '/spot/candlesticks'
    query_param = f'currency_pair={pair}&interval={interval}'
    if after != '':
        query_param += f'&from={after}'
    if before != '':
        query_param += f'&to={before}'

    r = requests.request('GET', host + prefix + url + "?" + query_param, headers=headers)
    return r.json()
```

```python
def get_current_timestamp():
    timestamp = int(time.time())
    hour = timestamp // 3600 * 3600
    return hour
```

```python
def get_all_candle_sticks(pair, start, interval='8h'):
    candle_sticks = []
    cur = get_current_timestamp()
    flag = True
    while start < cur:
        try:
            tmp = get_candle_sticks(pair, after=start, interval=interval)
            n_entries = len(tmp)
            print(n_entries)
            if 100 > n_entries > 0 and flag:
                flag = False
                tmp.pop()
            start += 100*8*3600
            candle_sticks += tmp
            time.sleep(0.2)
        except Exception as e:
            time.sleep(2)
    return candle_sticks
```

```python
import json

def get_pairs():

    # 读取 JSON 文件
    with open(f'{PATH}usdt_pairs.json', 'r') as file:
        json_data = file.read()

    # 将 JSON 数据解析为字典
    return json.loads(json_data)

    # 现在 data_dict 中包含了 JSON 文件中的数据，你可以使用它进行后续操作
```

```python
import os

def exist_file(file_name):
    return os.path.isfile(f'{PATH}{file_name}.csv')
```

```python
exist_file('AAVE_USDT')
```

```python
usdt_pairs = get_pairs()

for u_pair in usdt_pairs:
    if exist_file(u_pair):
        print(f'Exist {u_pair}')
        continue
    start_time = usdt_pairs[u_pair]
    if start_time == 0:
        print(f'Do not process {u_pair} now')
        continue
    data = get_all_candle_sticks(u_pair, start_time)
    df = pd.DataFrame(data)
    df.columns = ['timestamp', 'turnover', 'closing', 'high', 'low', 'open', 'volume']
    df.to_csv(f'{PATH}{u_pair}.csv', index=False)
    print(f'Finish {u_pair}!')
```
